```
import tensorflow as tf
from tensorflow import keras
import numpy as np

EPOCHS = 200
BATCH_SIZE = 128
VERBOSE = 1
NB_CLASSES = 10 
N_HIDDEN = 128
VALIDATION_SPLIT = 0.2

mnist = keras.datasets.mnist
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

print(len(X_train))
print(len(X_test))
print(tf.shape(X_train))
print(tf.shape(X_test))
RESHAPED = 784 #28*28

X_train = X_train.reshape(60000,RESHAPED)
X_test = X_test.reshape(10000,RESHAPED)
print(tf.shape(X_train))
print(tf.shape(X_test))

X_train = X_train.astype('float32') # 가중치 등에서 float 형태의 값들이 산출되기떄문에 int보다 float가 일반적인 형태
X_test = X_test.astype('float32')

#입력을 [0,1]사이로 정규화
X_train /= 255
X_test /=255
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')

Y_train = tf.keras.utils.to_categorical(Y_train, NB_CLASSES)
Y_test = tf.keras.utils.to_categorical(Y_test, NB_CLASSES)

model = tf.keras.models.Sequential()
#model.add(keras.layers.Dense(NB_CLASSES,
#                             input_shape=(RESHAPED,), 
#                             kernel_initializer='zeros', #가중치 0으로 초기화, random_uniform:-0.05~0.05사이 균등분포, random_normal : 평균0, 표준편차 0.05로 초기화
#                             name = 'dense_layer',
#                             activation='softmax'))
DROPOUT=0.3
model.add(keras.layers.Dense(N_HIDDEN,
                             input_shape = (RESHAPED,),
                             name='dense_layer', activation='relu'))
model.add(keras.layers.Dropout(DROPOUT))
model.add(keras.layers.Dense(N_HIDDEN,
                             name='dense_layer_2', activation='relu'))
model.add(keras.layers.Dropout(DROPOUT))
model.add(keras.layers.Dense(NB_CLASSES,
                             name='dense_layer_3', activation='softmax'))

model.compile(optimizer='SGD',
              loss = 'categorical_crossentropy',
              metrics = ['accuracy'])

model.fit(X_train, Y_train,
          batch_size = BATCH_SIZE, epochs=EPOCHS,
          verbose = VERBOSE, validation_split = VALIDATION_SPLIT)
```

```
test_loss, test_acc = model.evaluate(X_test, Y_test, batch_size=BATCH_SIZE)
print('Test accuracy : ', test_acc)
```
